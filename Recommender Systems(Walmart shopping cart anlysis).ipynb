```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Market_Basket_Optimisation.csv",header=None)
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
# Create a list of transactions

transactions=[]
```

```python
#There are a total of 7501 transacions
```

```python
#All transactions

for i in range(0,7501):

    #Adding Items for each transaction

    transactions.append([str(dataset.values[i,j]) for j in range(0,20)]) #Creates list of lists
```

```python
#Training Apriori on the dataset
from apyori import apriori 
```

```python
rules=apriori(transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2)

# support -> 3 items purchased 7 times a week divided by 7501 items

# confidence % of times our rule is correct
```

```python
#Visualising the results

results=list(rules)

output = []

for row in results:

    output.append(str(row.items))
```
